```python
from ultralytics import RTDETR
from PIL import Image
import pandas as pd
```

```python
import torch
print (torch.cuda.is_available())
```

```python
model = RTDETR("rtdetr-l.pt")
```

```python
results = model.train(
    data="halmet_dataset/data.yaml",
    epochs=100,
    imgsz=512,
    batch=8,
    device=0,
    workers=0,
    name="rtdetr"
    
)
```
